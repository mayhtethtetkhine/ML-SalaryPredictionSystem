```python
import numpy
import pandas
import joblib
import json
import seaborn

import sklearn

%matplotlib inline
```

```python
data_test = pandas.read_csv('./data/features.test.csv').set_index('id')
data_test
```

```python
final_data_test = pandas.read_csv('./data/features.finaltest.csv').set_index('id')
final_data_test
```

```python
model = joblib.load('model/model.joblib')
model
```

```python
data_test['prediction'] = model.predict(data_test.drop(['label'], axis='columns'))
data_test
```

```python
final_data_test['prediction'] = model.predict(final_data_test)
final_data_test
```

```python
data_test.to_csv('./results/predictions.test.csv')
```

```python
final_data_test.to_csv('./results/predictions.finaltest.csv')
```

```python
confusion_matrix_from_sklearn = sklearn.metrics.confusion_matrix(
    y_true = data_test['label'],
    y_pred = data_test['prediction'],
)
confusion_matrix_from_sklearn
```

```python
pandas.DataFrame(sklearn.metrics.classification_report(
    y_true = data_test['label'],
    y_pred = data_test['prediction'],
    digits = 4,
    output_dict = True,
))
```
