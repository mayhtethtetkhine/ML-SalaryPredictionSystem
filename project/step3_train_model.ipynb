```python
import joblib
import json
import pandas

import sklearn.neural_network
import warnings
warnings.filterwarnings('ignore')
```

```python
data_train = pandas.read_csv('./data/features.train.csv').set_index('id')
data_train
```

```python
data_validation = pandas.read_csv('./data/features.test.csv').set_index('id')
data_validation
```

```python
model = sklearn.neural_network.MLPClassifier(

    hidden_layer_sizes    = (17,8, 17),
    solver                = 'sgd', 
    activation            = 'logistic', 
    alpha                 = 0.0029,
    batch_size            = 36, 
    learning_rate         = 'constant',
    learning_rate_init    = 0.12,
    max_iter              = 1000, 
    momentum              = 0.0, 
    nesterovs_momentum    = False, 
    validation_fraction   = 0.2, 
    shuffle               = True,
    random_state          = 0

)

model
```

```python
model.fit(
    X = data_train.drop(columns=['label']),
    y = data_train['label']
)
```

data_train.drop(columns=['label'])

```python
 joblib.dump(model, './model/model.joblib')
```

```python
with open('./model/config.json', 'w') as f: 
    json.dump(model.get_params(), f, indent = 4)
```
